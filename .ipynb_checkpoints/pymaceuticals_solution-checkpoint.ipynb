```python
# SUMMARY STATISTICS
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# This method is the most straighforward, creating multiple series and putting them all together at the end.
 
regimen_grouped = clean_df.groupby(["Drug Regimen"])

# Applying aggregates to the groupby 
mean_1 = regimen_grouped["Tumor Volume (mm3)"].mean()
med_1 = regimen_grouped["Tumor Volume (mm3)"].median()
var_1 = regimen_grouped["Tumor Volume (mm3)"].var()
sd_1 = regimen_grouped["Tumor Volume (mm3)"].std()
sem_1 = regimen_grouped["Tumor Volume (mm3)"].sem()

# Generating a summary statistics table to hold results

tumorvol_summary1 = pd.DataFrame({
    "Mean Tumor Volume (mm3)" : mean_1,
    "Median Tumor Volume (mm3)" : med_1,
    "Variance Tumor Volume (mm3)" : var_1,
    "Standard Deviation Tumor Volume (mm3)" : sd_1,
    "SEM Tumor Volume (mm3)" : sem_1
})
tumorvol_summary1
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen
# This method produces everything in a single groupby function

tumorvol_summary2 = regimen_grouped["Tumor Volume (mm3)"].agg(["mean", "median", "var", "std", "sem"])              
# Renaming column headers so looks the same as tumorvol_summary1
tumorvol_summary2.rename(columns={
    "mean":"Mean Tumor Volume (mm3)",
    "median":"Median Tumor Volume (mm3)",
    "var":"Variance Tumor Volume (mm3)",
    "std":"Standard Deviation Tumor Volume (mm3)",
    "sem":"SEM Tumor Volume (mm3)"
})

tumorvol_summary_2
```

```python
# Grouping the clean data by mouse ID and then calculating max timepoint
mouse_grouped = clean_df.groupby(["Mouse ID"])
max_tp = mouse_grouped["Timepoint"].max()
max_tp
```

```python
# Merging this to the original datafrtame
merged_df = pd.merge(clean_df, max_tp, on='Mouse ID')

# Now there's 2 timepoints_x and _y, rename
merged_df = merged_df.rename(columns={"Timepoint_x" : "Timepoint", "Timepoint_y": "Last Timepoint"})
merged_df
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ["Capomulin", "Ramicane", "Infubinol", "Ceftamin"]
```

```python
# Create empty list to fill with tumor vol data (for plotting)
tumorvol_data = []
```

```python
# for loop to get the final tumor volume for each of 4 treatments

for treatment in treatments:

    # Locing in on the 4 treatments in list
    treat_df = merged_df.loc[merged_df["Drug Regimen"] == treatment]

    # Locing in on timepoint where it equals the final timepoint
    lastvol_df = treat_df.loc[treat_df["Timepoint"] == treat_df["Last Timepoint"]]
    
    # Pulling out the tumor volume for the above
    tumor_vol = lastvol_df['Tumor Volume (mm3)']
    
    # Appending these values to my empty list
    tumorvol_data.append(tumor_vol)
    
    # Calculating the IQR
    # If the data is in a dataframe, we use pandas to give quartile calculations
    quartiles = tumor_vol.quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq - lowerq
    
    # Determining outliers using upper and lower bounds
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    
    # Print findings
    print(f"For {treatment}, the interquartile range of final tumor volume is {iqr.round(2)}(mm3)")
    print(f"For {treatment} values below {lower_bound.round(2)}(mm3) could be outliers")
    print(f"For {treatment} values above {upper_bound.round(2)}(mm3) could be outliers")
```

```python
# Giving potential outliers custom style or color according to grading rubric
red_diamond = dict(markerfacecolor='r', marker='D')

# Generating box plot
fig1, ax1 = plt.subplots()
ax1.set_title('Final Tumor Volume showing potential outlier')
ax1.set_xlabel('Treatment type')
ax1.set_ylabel('Final Tumor Volume (mm3)')
ax1.boxplot(tumorvol_data, flierprops = red_diamond)
plt.show()
```

```python
# Generate a scatter plot of mouse weight versus average tumor volume for the Capomulin regimen

# Loc in on where regimen = capomulin
spec_drug = clean_df.loc[clean_df["Drug Regimen"] == "Capomulin"]
spec_drug.head()
```

```python
# Calculating average tumor volume
cap_ID = spec_drug.groupby(["Mouse ID"])
mean_vol = cap_ID["Tumor Volume (mm3)"].mean()
merged = pd.merge(spec_drug, mean_vol, on='Mouse ID')
merged = merged.rename(columns={"Tumor Volume (mm3)_x" : "Tumor Volume (mm3)" , "Tumor Volume (mm3)_y" : "Average Tumor Volume (mm3)"})
scatter_df = merged[['Weight (g)', 'Average Tumor Volume (mm3)']].drop_duplicates()
scatter_df.head()
```

```python
x_scatter = scatter_df["Weight (g)"]
y_scatter = scatter_df["Average Tumor Volume (mm3)"]
plt.scatter(x_scatter, y_scatter, marker="o", facecolors="red", edgecolors="black")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.title("Mouse weight vs average tumor volume for mice treated with Capomulin")
plt.show()
```

```python
# Correlation and Regression
# Calculate the correlation coefficient and linear regression model for mouse weight and average tumor volume for the Capomulin regimen
```

```python
correlation = st.pearsonr(x_scatter,y_scatter)
print(f"The correlation coefficient between mouse weight and average tumor volume for the Capomulin regimen is {correlation[0].round(2)}")
```

```python
# Linear regression model
x_values = x_scatter
y_values = y_scatter
(slope, intercept, rvalue, pvalue, stderr) = st.linregress(x_values, y_values)
regress_values = x_values * slope + intercept
line_eq = "y = " + str(round(slope,2)) + "x + " + str(round(intercept,2))
plt.scatter(x_values,y_values)
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(0,50),fontsize=15,color="red")
plt.xlabel("Weight (g)")
plt.ylabel("Average Tumor Volume (mm3)")
plt.title("Mouse weight vs average tumor volume for mice treated with Capomulin")
print(f"The r-squared is: {rvalue**2})")
plt.show()
```
